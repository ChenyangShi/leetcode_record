#### 数据总和加起来小于等于k的最长的连续子数组

```python
import sys
import numpy as np

arr = [1, 1, 1, 9, 9, 1, 1, 1, 1]
#arr = [1, 2, 1, 3, 1, 1, 1]
k = 4

sum = np.zeros(len(arr), dtype=int)
length = np.zeros(len(arr), dtype=int)

if k>=2:
    sum[0] = arr[0]
    length[0] = 1
    for i in range(1, len(arr)):
        sum[i] = arr[i] + sum[i-1]
        if sum[i] > 4:
            s = length[i-1]
            sum[i] = sum[i] - arr[i-s]
            length[i] = length[i-1]
        else:
            length[i] = length[i-1] + 1
            
else:
    print("no")
    
print(sum)
print(length)

    
sum = [x for x in sum if x <= 4]
length = [x for x in length if x <= 4]
print(sum)
print(length)

res = max(length)
print(res)
```
